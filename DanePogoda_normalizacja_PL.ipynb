```python
import pandas as pd
```

```python
# funkcja do kategoryzacji kolumny Godzina
def categorize_time(time_str):
    hour = int(time_str.split(':')[0])
    if hour >= 21 or hour < 5:
        return 'noc'
    elif hour >= 6 and hour <= 16:
        return 'rano'
    else:
        return 'wieczór'
```

```python
df = pd.read_csv("DanePogoda.csv", encoding= 'unicode_escape')
df.head()
```

```python
for name, dtype in df.dtypes.items():
    print(name, dtype)
```

```python
zachmurzenie_dict = {
    0: "Brak",
    1: "małe",
    2: "małe",
    3: "małe",
    4: "umiarkowane",
    5: "umiarkowane",
    6: "duże",
    7: "duże",
    8: "pełne",
}

df.Zachmurzenia = [zachmurzenie_dict[item] for item in df.Zachmurzenia]
df.head(10)
```

```python
df['Godzina'] = df['Godzina'].apply(categorize_time)

df.head(10)
```

```python
df.head(30)
```

```python
# Konwertuj zmienną kategoryczną na zmienne fikcyjne/wskaźnikowe.
# Każda zmienna jest konwertowana na tyle zmiennych 0/1, ile jest różnych wartości.
# Kolumny w danych wyjściowych są nazwane według wartości; jeśli dane wejściowe to DataFrame,
# nazwa oryginalnej zmiennej jest dodawana do wartości.
one_hot_df_godzina = pd.get_dummies(df['Godzina'], prefix='')
one_hot_df_zachmurzenie = pd.get_dummies(df['Zachmurzenia'], prefix='')
one_hot_df_zjawisko = pd.get_dummies(df['zjawisko'], prefix='')

# Połącz obiekty pandy wzdłuż określonej osi.
# Umożliwia opcjonalną logikę zestawu wzdłuż innych osi.
# Można również dodać warstwę indeksowania hierarchicznego na osi konkatenacji, co może być przydatne,
# jeśli etykiety są takie same (lub nakładają się) na przekazywanym numerze osi.
# polecenie concat służy nam do dołączenia zakodowanych wcześniej danych do oryginalnego zestawu danych
df = pd.concat([df, one_hot_df_godzina], axis=1)
df = pd.concat([df, one_hot_df_zachmurzenie], axis=1)
df = pd.concat([df, one_hot_df_zjawisko], axis=1)

df.head(5)
```

```python
# Znajduję minimalną i maksymalną wartość kolumny
min_cisnienie = df['Cisnienie'].min()
max_cisnienie = df['Cisnienie'].max()

# Normalizuję kolumnę 
df['Cisnienie'] = (df['Cisnienie'] - min_cisnienie) / (max_cisnienie - min_cisnienie)

df.head(10)
```

```python
df = df.drop(columns=['Godzina', 'zjawisko','Zachmurzenia'])
df.head(30)
```
